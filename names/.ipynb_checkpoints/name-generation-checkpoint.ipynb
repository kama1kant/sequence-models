```python
%matplotlib inline
from __future__ import unicode_literals, print_function, division
from io import open
import glob
import os
import unicodedata
import string
import torch
import torch.nn as nn
import random
import time
import math
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
class Util():
    def __init__(self, classifier):
        pass
        self.all_letters = string.ascii_letters + " .,;'"
        self.n_letters = len(self.all_letters)

    def randomChoice(self, l):
        return l[random.randint(0, len(l) - 1)]

    def randomTrainingPair(self):
        category = self.randomChoice(classifier.all_categories)
        line = self.randomChoice(classifier.category_lines[category])
        return category, line
    
    def categoryTensor(self, category):
        li = classifier.all_categories.index(category)
        tensor = torch.zeros(1, classifier.n_categories)
        tensor[0][li] = 1
        return tensor

    def inputTensor(self, line):
        tensor = torch.zeros(len(line), 1, self.n_letters)
        for li in range(len(line)):
            letter = line[li]
            tensor[li][0][self.all_letters.find(letter)] = 1
        return tensor

    def targetTensor(self, line):
        letter_indexes = [self.all_letters.find(line[li]) for li in range(1, len(line))]
        letter_indexes.append(self.n_letters - 1) # EOS
        return torch.LongTensor(letter_indexes)
    
    def findFiles(self, path): return glob.glob(path)
    
    def unicodeToAscii(self, s):
        return ''.join(
            c for c in unicodedata.normalize('NFD', s)
            if unicodedata.category(c) != 'Mn'
            and c in self.all_letters
        )
    
    def readLines(self, filename):
        lines = open(filename, encoding='utf-8').read().strip().split('\n')
        return [self.unicodeToAscii(line) for line in lines]

    def timeSince(self, since):
        now = time.time()
        s = now - since
        m = math.floor(s / 60)
        s -= m * 60
        return '%dm %ds' % (m, s)
    
    def categoryFromOutput(self, output):
        top_n, top_i = output.topk(1)
        category_i = top_i[0].item()
        return classifier.all_categories[category_i], category_i
    
    def randomTrainingExample(self):
        category, line = self.randomTrainingPair()
        category_tensor = self.categoryTensor(category)
        input_line_tensor = self.inputTensor(line)
        target_line_tensor = self.targetTensor(line)
#         print(category, line)
#         print(category_tensor, input_line_tensor, target_line_tensor)
#         print(util.all_letters)
        return category_tensor, input_line_tensor, target_line_tensor
```

```python
class RNN(nn.Module):
    def __init__(self, classifier, input_size, hidden_size, output_size):
        super(RNN, self).__init__()
        self.hidden_size = hidden_size

        self.i2h = nn.Linear(classifier.n_categories + input_size + hidden_size, hidden_size)
        self.i2o = nn.Linear(classifier.n_categories + input_size + hidden_size, output_size)
        self.o2o = nn.Linear(hidden_size + output_size, output_size)
        self.dropout = nn.Dropout(0.1)
        self.softmax = nn.LogSoftmax(dim=1)

    def forward(self, category, input, hidden):
        input_combined = torch.cat((category, input, hidden), 1)
        hidden = self.i2h(input_combined)
        output = self.i2o(input_combined)
        output_combined = torch.cat((hidden, output), 1)
        output = self.o2o(output_combined)
        output = self.dropout(output)
        output = self.softmax(output)
        return output, hidden

    def initHidden(self):
        return torch.zeros(1, self.hidden_size)
```

```python
class Name_Classifier():
    def __init__(self):
        self.util = Util(self)
        self.init_hparams()
        self.getData()
        self.getModel()
        
    def init_hparams(self):
        self.n_iters = 100000
        self.print_every = 5000
        self.plot_every = 1000
        self.criterion = nn.NLLLoss()
        self.learning_rate = 0.005
        self.n_hidden = 128
        self.total_loss = 0
        self.max_length = 20
    
    def getData(self):
        self.category_lines = {}
        self.all_categories = []

        for filename in self.util.findFiles('data/names/*.txt'):
            self.category = os.path.splitext(os.path.basename(filename))[0]
            self.all_categories.append(self.category)
            self.lines = self.util.readLines(filename)
            self.category_lines[self.category] = self.lines

        self.n_categories = len(self.all_categories)
        if self.n_categories == 0:
            raise RuntimeError('Data not found. Make sure that you downloaded data '
        'from https://download.pytorch.org/tutorial/data.zip and extract it to '
        'the current directory.')
    
    def getModel(self):
        self.rnn = RNN(self, self.util.n_letters, self.n_hidden, self.util.n_letters)
        
    def train(self, category_tensor, input_line_tensor, target_line_tensor):
        target_line_tensor.unsqueeze_(-1)
        hidden = self.rnn.initHidden()

        self.rnn.zero_grad()

        loss = 0

        for i in range(input_line_tensor.size(0)):
            output, hidden = self.rnn(category_tensor, input_line_tensor[i], hidden)
            print(output.shape, target_line_tensor.shape, target_line_tensor[i].shape)
            l = self.criterion(output, target_line_tensor[i])
            loss += l

        loss.backward()

        for p in self.rnn.parameters():
            p.data.add_(p.grad.data, alpha=-self.learning_rate)

        return output, loss.item() / input_line_tensor.size(0)
    
    def fit(self):
        self.all_losses = []
        start = time.time()
        
        for iter in range(1, self.n_iters + 1):
            output, loss = self.train(*util.randomTrainingExample())
            self.total_loss += loss

            if iter % self.print_every == 0:
                print('%s (%d %d%%) %.4f' % (util.timeSince(start), iter, iter / self.n_iters * 100, loss))

            if iter % self.plot_every == 0:
                self.all_losses.append(self.total_loss / self.plot_every)
                self.total_loss = 0
                
    def sample(self, category, start_letter='A'):
        with torch.no_grad():
            category_tensor = self.util.categoryTensor(category)
            input = self.util.inputTensor(start_letter)
            hidden = self.rnn.initHidden()

            output_name = start_letter

            for i in range(self.max_length):
                output, hidden = self.rnn(category_tensor, input[0], hidden)
                topv, topi = output.topk(1)
                topi = topi[0][0]
                if topi == self.util.n_letters - 1:
                    break
                else:
                    letter = self.util.all_letters[topi]
                    output_name += letter
                input = self.util.inputTensor(letter)

            return output_name

    def samples(self, category, start_letters='ABC'):
        for start_letter in start_letters:
            print(self.sample(category, start_letter))
    
    def plot_loss(self):
        plt.figure()
        plt.plot(self.all_losses)
```

```python
classifier = Name_Classifier()
util = Util(classifier)
```
